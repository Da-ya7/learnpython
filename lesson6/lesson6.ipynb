**return** — *sending a value BACK out of the function*

Big distinction: print() just shows something on screen. return actually hands a value back so you can use it elsewhere in your code.

```python
def add(a,b):
    return a + b

result=add(3,5)
print(result)
```

Here, add(3, 5) runs, computes 8, and gives it back — it gets stored in the result box. Without return, the function would compute 8 and then just... throw it away, doing nothing useful.

**comparing print and retrun**

```python
def add_print(a, b):
    print(a + b)   # just displays it, gives nothing back

def add_return(a, b):
    return a + b   # hands the value back

x = add_print(3, 5)   # prints 8, but x is None (nothing returned!)
y = add_return(3, 5)  # nothing printed, but y = 8

print(x)   # None
print(y)   # 8
```

remember it: print shows, return gives back.

**A function stops immediately at return**

```python
def check_age(age):
    if age < 0:
        return "Invalid age"
    return "valid age"
    print("This never runs!")  # unreachable — return already exited the function
check_age(-2)
```

***Default parameter values:***
*You can give a parameter a "default" value to use if the caller doesn't provide one.*


```python
def greet(name="Friend"):
    print(f"hello {name} !")

greet()
greet("Daya")
```

*args — accepting an unknown number of arguments 

```python
def add_all(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total
print(add_all(1,2,3))
print(add_all(1, 2, 3, 4, 5))
```

*args → packs positional arguments into a tuple.


**kwargs → packs keyword arguments into a dictionary.

**kwargs — accepting unknown named arguments

```python
def describe(**info):
    for key,value in info.items():
        print(key+ ":" ,value)

describe(name="Alex", age=10,city="Delhi")
```

****info.items() gives the for loop one (key, value) tuple at a time.****
