**The main tool: try / except**

```python
try:
    age=int(input("Please enter your age: "))
    print("Next year you will be",age+1)
except:
    print("Enter a valid number")
```

*Being specific about WHICH error you catch*

```python
try:
    age=int(input("Enter your age:"))
    print("next year you will be",age+1)
except ValueError:
    print("thats not a number")
```

*Catching multiple different errors*

```python
try:
    number=int(input("Enter the number"))
    result=10/number
    print(result)
except ValueError:
    print("Thats not a number")
except ZeroDivisionError:
    print("Number cannot divided by zero")
```

*else — runs only if NO error happened*

```python
try:
    number=int(input("Enter the number:"))
except ValueError:
    print("This is not a number")
else:
    print("Great you entered a number")
```

The else block only runs if the try block succeeded completely, with zero errors.

**finally — ALWAYS runs, error or not**

```python
try:
    number=int(input("Enter the number:"))
except ValueError:
    print("This is not a number")
finally:
    print("This will run no matter what")
```

**Grabbing the actual error message**---->
*Sometimes you want to see what Python is actually complaining about, not just handle it silently:*

```python
try:
    number=10/0
except ZeroDivisionError as e:
    print("Error occured",e)
```

**Raising your own errors: raise**--->>>
*Sometimes you want to stop the program and throw an error deliberately, because something doesn't make sense for your program's logic.*

```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```
