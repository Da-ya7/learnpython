*The building blocks: if, elif, else*

```python
weather="rainy"

if weather == "rainy":
    print("Take an umbrella")
elif weather == "sunny":
    print("Wear sunglasses")
else:
    print("Just go outside normally")
```

*Combining conditions: and, or, not*

```python
age=20
has_id=True

if age >=18 and has_id:
    print("You can entre")
else:   
    print("you cannot enter")
```
