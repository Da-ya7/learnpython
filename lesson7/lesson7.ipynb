***for loops — when you know what you're looping over***

```python
fruits=["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

***Looping over text (strings)***

```python
for letter in "Leon kennedy":
    print(letter, end="")
```

****range() — looping a specific number of times****

```python
for i in range(5):
    print(i, end="")
```

**********Very common pattern: using range(len(list)) to get positions in a list instead of the items themselves:**********


```python
fruits = ["apple","banana","cherry"]
for i in range(len(fruits)):
    print(i,"-",fruits[i])
```

few example problems

```python
numbers=[10,20,30,40,50]

for i in range(len(numbers)):
    print(i,"-",numbers[i])
```

```python
names=["Ram","Sita","Krishna"]

for i in range(1,len(names)):
    print(i,"-",names[i])
```

```python
numbers=[10,20,30,40,50]

for i in range(2):
    print(i,"=",numbers[i])
```

```python
letters=["A","E","I","O","U"]

for i in range(1,5,2):
    print(i,":",letters[i])
```

***enumerate() gives both the index and the value while looping through a list.***

```python
fruits = ["apple", "banana", "cherry"]

for i, fruit in enumerate(fruits):
    print(i, "-", fruit)
```

```python
animals = ["Dog", "Cat", "Lion"]

for a, b in enumerate(animals):
    print(a)
    print(b)
```

***while loops — when you don't know how many times, only WHEN to stop***

A while loop keeps running as long as a condition stays True. You use this when you don't know in advance how many repetitions you'll need.

```python
count =0

while count<5:
    print(count, end=" ")
    count+=1
```

**while True — loop forever on purpose, until you break it**

```python
while True:
    answer=input("Enter 'quit' to stop:")
    if answer == 'quit':
        break
    print("You typed:",answer)
```

break — exit a loop immediately

```python
for i in range(10):
    if i==5:
        break
    print(i)
```

continue — skip the rest of THIS round, go to the next one

```python
for i in range(5): 
    if i == 2:
        continue
    print(i)
```

**Nested loops — a loop inside a loop**

```python
for i in range(3):
    for j in range(2):
        print(i,j)
```

**else on a loop (uncommon, but good to know)**

The else runs only if the loop completed without hitting a break. Rarely used, but you'll see it occasionally.

```python
for i in range(5):
    print(i)
else:
    print("Loop finished normally")
```
