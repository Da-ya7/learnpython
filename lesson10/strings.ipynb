**Working with Strings**

A string is just text — a sequence of characters wrapped in quotes. But under the hood, Python treats it like a list of characters, which means a LOT of the list tricks you already learned apply here too.

```python
name="Alex"
```

You can use single '...' or double "..." quotes — no real difference, just be consistent.

***Strings are indexed, just like lists***

```python
word="Python"
print(word[0])
print(word[-6])
print(word[-5])
```

***Slicing strings — exactly like list slicing***

```python
word = "Python"
print(word[0:3])   # Pyt   (index 0 up to, not including, 3)
print(word[2:])    # thon
print(word[:2])    # Py
print(word[::-1])  # nohtyP  <- reverses the whole string! (step = -1)
```

That last one, [::-1], is a classic trick: start:stop:step — leaving start/stop empty means "whole string," and step -1 means "walk backwards." Super common way to reverse a string.

Strings are IMMUTABLE (can't change them directly)

```python
word = "Python"
word[0] = "J"   # ❌ ERROR — TypeError: 'str' object does not support item assignment
```

```python
word = "Python"
word = "J" + word[1:]   # "Jython" -- built a new string, didn't edit the old one
print(word)
```

len()

```python
name='Alex'
print(len(name))
```

Concatenation — joining strings with +

```python
first_name="Ajith"
last_name="Kumar"
full_name=first_name+" "+last_name
print(full_name)
```

Watch out — you can't mix strings and numbers directly with +:

```python
age = 10
print("Age: " + age)         # ❌ ERROR
print("Age: " + str(age))    # ✅ "Age: 10"
```

f-strings — the MODERN, BEST way to build strings with variables

This is what you should actually use in real code, instead of + chains:

```python
name = "Alex"
age = 10
print(f"My name is {name} and I am {age} years old")
```

```python
# You can even do math/logic INSIDE the braces:
price = 19.99
print(f"Total with tax: {price * 1.08:.2f}")
# the ":.2f" formats it to 2 decimal places -> Total with tax: 21.59
```

****Common string methods****

```python
text=" Hello Learners "
print(text.strip()) #removes leading/trailing whitespace
print(text.lower()) #-- all lowercase
print(text.upper()) #-- all uppercase
print(text.title()) #Capitalizes Each Word
print(text.replace("Learners","Python"))
```

None of these change text itself (remember — strings are immutable!). They all return a new string. If you want to keep the change, you must reassign it:

```python
text = text.strip()   # NOW text is actually updated
```

Splitting and joining — converting between strings and lists

```python
sentence = "the quick brown fox"
words = sentence.split()          # split by whitespace (default)
print(words)   # ['the', 'quick', 'brown', 'fox']

csv_line = "apple,banana,cherry"
items = csv_line.split(",")       # split by a specific character
print(items)   # ['apple', 'banana', 'cherry']

# The reverse: join a list back into a string
words = ['the', 'quick', 'brown', 'fox']
sentence = " ".join(words)         # glue them together with " " between each
print(sentence)   # "the quick brown fox"
```

The split() method converts a string into a list by splitting it wherever a separator occurs.

***.split() and .join() are used everywhere — parsing text input, reading CSV-style data, building URLs, etc. Very important for web dev.***

*************************Checking / searching inside strings*************************

```python
text = "Hello World"

print("World" in text)          # True — substring check
print(text.startswith("Hello")) # True
print(text.endswith("World"))   # True
print(text.find("World"))       # 6  -- index where it starts (returns -1 if not found)
print(text.count("l"))          # 3  -- how many times "l" appears
```

Checking what KIND of characters a string has

```python
"123".isdigit()      # True -- all characters are digits
"abc".isalpha()       # True -- all characters are letters
"abc123".isalnum()    # True -- letters and/or numbers only
"   ".isspace()        # True -- all whitespace
```

Handy for validating user input — e.g., checking if someone typed a valid number before you try int() on it (ties right back to what you learned in Exceptions

String multiplication (a fun quirk)

```python
print("ha" * 3)     # "hahaha"
print("-" * 20)      # "--------------------" -- great for dividers/separators
```

Putting it together — a realistic mini example

```python
raw_input = "  Alex Kumar, 25, Chennai  "

cleaned = raw_input.strip()              # remove outer whitespace
parts = cleaned.split(",")                # split into pieces by comma
name = parts[0].strip()
age = parts[1].strip()
city = parts[2].strip()

print(f"Name: {name}, Age: {age}, City: {city}")
```
