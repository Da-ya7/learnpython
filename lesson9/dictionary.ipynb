****A dictionary stores information as key-value pairs — instead of finding things by position (like a list), you find things by name (key).****

```python
person={
    "name":"Alex",
    "age":10,
    "city":"Delhi"
}
```

Curly brackets { }, with key: value pairs separated by commas.

**Accessing values — by key, not by position**

```python
print(person["name"])
print(person["age"])
print(person["city"])
```

There's no index 0 here — dictionaries aren't ordered by position, they're organized by key name.

*****************Safer way — .get() — won't crash, just gives you None (or a default you choose) if the key is missing:*****************

```python
print(person.get("height"))            # None
print(person.get("height", "unknown")) # "unknown"  -- you choose the fallback
```

Always prefer .get() when you're not 100% sure a key exists — this alone will save you from a LOT of crashes.

****Adding and updating values****

```python
person={"name":"Alex","age":10}
person["age"]=20
person["city"]="Delhi"
print(person)
```

Same syntax for both adding and updating — Python checks: does this key already exist? If yes, update it. If no, create it.

***Removing items***

```python
person = {"name": "Alex", "age": 10, "city": "Delhi"}

del person["city"]         # removes the key entirely
print(person)               # {'name': 'Alex', 'age': 10}

age = person.pop("age")     # removes AND gives you back the value
print(age)                   # 10
print(person)                # {'name': 'Alex'}
```

*****Checking if a key exists*****

```python
person = {"name": "Alex", "age": 10}
print("name" in person)     # True — checks KEYS by default
print("Alex" in person)     # False — "Alex" is a value, not a key!
print("Alex" in person.values()) #look for values 
print(("name", "Alex") in person.items())# checks for Key-value pairs
```

****Looping through a dictionary — 3 waysA****

```python
person = {"name": "Alex", "age": 10, "city": "Delhi"}

# 1. Loop over keys (default behavior)
for key in person:
    print(key)
# name, age, city

# 2. Loop over values only
for value in person.values():
    print(value)
# Alex, 10, Delhi

# 3. Loop over BOTH key and value together (most useful, most common)
for key, value in person.items():
    print(key, ":", value)
# name : Alex
# age : 10
# city : Delhi
```

Nested dictionaries — dictionaries inside dictionaries
Very common in real projects (e.g., representing a database of multiple people):

```python
students={
    "alex":{"age":10,"grade":5},
    "sam":{"age":11,"grade":6}
}
print(students["alex"])
print(students["alex"]["age"])
for students_name,info in students.items():
    print(students_name,"is",info["age"],"years old")
```
