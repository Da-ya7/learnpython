**1. Lists — the flexible, changeable collection**

****What is it?****--
A list is an ordered, changeable collection of items. Think of it like a shopping list written in pencil — you can add items, cross them out, rearrange them, whatever you want.

```python
fruits=["apple", "banana", "cherry"]
print(fruits)
```

***Accessing items — indexing :***
Every item has a position number, starting at 0 (not 1!).

```python
fruits=["apple","banana","cherry"]
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

***Negative indexing*** — *counting from the end*

```python
print(fruits[-1])
print(fruits[-2])
print(fruits[-3])
```

*Think of -1 as "the last one," -2 as "second from last," and so on.*

Slicing — grabbing a chunk of the list

```python
numbers = [10, 20, 30, 40, 50]
print(numbers[1:3])    # [20, 30]  -- from index 1 up to (not including) 3
print(numbers[:2])     # [10, 20]  -- from start up to index 2
print(numbers[2:])     # [30, 40, 50]  -- from index 2 to the end
print(numbers[:])      # [10, 20, 30, 40, 50] -- a full copy
```

Rule: list[start:stop] → includes start, stops before stop.

****Changing a list (this is the key thing that makes it a "list" and not a tuple)****

```python
fruits=["apple","banana","cherry"]
fruits[0]="mango"
print(fruits)
```

Common list methods (built-in tools for lists)

```python
fruits = ["apple", "banana"]

fruits.append("cherry")        # add to the end       -> ['apple','banana','cherry']
fruits.insert(1, "kiwi")       # insert at position 1  -> ['apple','kiwi','banana','cherry']
fruits.remove("banana")        # remove by value       -> ['apple','kiwi','cherry']
fruits.pop()                   # remove & return LAST item
fruits.pop(0)                  # remove & return item at index 0
fruits.sort()                  # sort alphabetically/numerically, in place
fruits.reverse()                # flip the order, in place
print(len(fruits))              # how many items
print("kiwi" in fruits)         # True/False — is it in the list?
fruits.clear()                  # empty the whole list
```

append vs insert: append always adds to the end, insert(position, value) lets you choose where.

= creates a reference, not a copy.
