*****Tuples — like a list, but LOCKED*****

*****What is it?*****
A tuple is an ordered, but UNCHANGEABLE (immutable) collection. Once created, you cannot add, remove, or change items.

```python
point=(10,20)
print(point[0])
print(point[1])
```

Round brackets ( ) = tuple.

Why would you want something you can't change?

Because sometimes that's exactly the point — you want to guarantee the data can't be accidentally messed with elsewhere in your program. Example: GPS coordinates, RGB color values, days of the week — things that shouldn't change once set.

```python
colour=(255,0,0)  # RGB for red — this shouldn't accidentally change
```

*****Trying to change a tuple = error*****

```python
point=(10,20)
point[0]=99
```

***What We CAN do with tuples***

```python
point=(10,20,30)
print(len(point))
print(point[1:3])
print(20 in point)
for value in point:
    print(value)
```

***Tuple unpacking — a neat trick you'll use a LOT***

```python
point=(10,20,30)
x,y ,z=point
print(x)
print(y)
print(z)
```

*****This is everywhere in real Python code — e.g., getting a function's multiple return values:*****

```python
def get_dimensions():
    return (1920, 1080)   # returning a tuple

width, height = get_dimensions()
print(width)    # 1920
print(height)   # 1080
```

***One-item tuple — a sneaky trap***

```python
not_a_tuple = (5)      # this is just an int in parentheses! type is int
actual_tuple = (5,)    # the COMMA is what makes it a tuple
print(type(not_a_tuple))   # <class 'int'>
print(type(actual_tuple))  # <class 'tuple'>
```
