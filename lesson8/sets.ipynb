****Sets — unordered, no duplicates allowed****

A set is a collection where order doesn't matter and duplicates are automatically removed. Think of it like a bag of unique stickers — you can't have two identical ones, and there's no "first" or "second" sticker in a bag.

```python
number={1,2,3,1,2}
print(number)
```

Curly brackets { } = set (but watch out — empty {} is actually a dictionary, not a set! Empty set is set()).

***Why use a set?***

Removing duplicates from a list, instantly:

```python
names = ["Alex", "Sam", "Alex", "Jo", "Sam"]
unique_names = set(names)
print(unique_names)   # {'Alex', 'Sam', 'Jo'}
```

***********Super fast membership checking (in on a set is much faster than in on a big list)***********

```python
allowed_users = {"alex", "sam", "jo"}
print("alex" in allowed_users)   # True — checked almost instantly, even for huge sets
```

***No indexing! (because there's no order)***

```python
numbers = {1, 2, 3}
print(numbers[0])   # ❌ ERROR — sets aren't ordered, no index 0 exists
```

*****Common set methods*****

```python
a = {1, 2, 3}
a.add(4)          # add one item        -> {1,2,3,4}
a.remove(2)        # remove an item (errors if missing)
a.discard(99)       # remove if present, NO error if missing
```

**********************************Set math — this is where sets really shine**********************************

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union: everything in either       -> {1,2,3,4,5,6}
print(a & b)   # intersection: in BOTH             -> {3,4}
print(a - b)   # difference: in a but NOT in b      -> {1,2}
print(a ^ b)   # symmetric difference: in one, not both -> {1,2,5,6}
```
